Inisialisasi

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

# Inisialiasi keterangan variabel pada rumus
x = 0                                   # inisialisasi nilai x
y = 0                                   # inisialisasi nilai y
g = 9.806                               # nilai kecepatan gravitasi
ax = 0                                  # rumus 1 : percepatan pada sumbu x
ay = -g                                 # rumus 2 : percepatan pada sumbu y
m = 0.15                                # massa
v0 = 50                                 # nilai kecepatan dari soal 
sudut = 35                              # sudut derajat
sudut_rad = (sudut/360)*(2 * np.pi)     # rumus mengubah sudut menjadi radian
d = 0.0013                              # hambatan udara
dt = 0.01                               # time step(delta t)
t = 0                                   # waktu

# Array untuk menampung setiap perhitungan numerik
array_x = [x]                             # untuk menampung setiap hasil perhitungan numerik mengenai jarak
array_y = [y]                             # untuk menampung setiap hasil perhitungan numerik mengenai ketinggian
array_t = [t]                             # untuk menampung setiap hasil perhitungan numerik mengenai waktu

vx = v0 * np.cos(sudut_rad)               # rumus 7 : v(0) cos alfa (menghitung kecepatan awal pada sumbu x)
vy = v0 * np.sin(sudut_rad)               # rumus 8 : v(0) sin alfa (menghitung kecepatan awal pada sumbu y)
```

#### Solusi Numerik Tanpa Hambatan

```python
# melakukan perulangan untuk perhitungan numerik dengan waktu(t) yang berbeda tiap perulangan
while y >= 0 :                                
  vx += ax*dt                                 # rumus 5 : vx(t + delta t) = vx(t) + (ax + delta t) "Konstan, jawaban tidak akan berubah meskipun perulangan beberapa kali"
  vy += ay*dt                                 # rumus 6 : vx(t + delta t) = vy(t) + (ay + delta t) 
  x += vx * dt                                # rumus 3 : x(t + delta t) = x(t) + vx(t + delta t) * delta t
  y += vy * dt                                # rumus 4 : y(t + delta t) = y(t) + vy(t + delta t) * delta t
  t += dt
  if y < 0 :                                  
    break                                     
  # menampung hasil perhitungan numerik yang telah dilakukan
  array_x.append(x)
  array_y.append(y)
  array_t.append(t)

print("Hasil Perhitungan Metode Numerik")
print()

#waktu total perhitungan numerik
ttot_numerik = array_t[-1]                                                  
print("Waktu total perhitungan numerik          : ", ttot_numerik, " s")

#jarak maksimum perhitungan numerik
rmax_numerik = array_x[-1]                                                  
print("Jarak maksimum perhitungan numerik       : ", rmax_numerik, " m")

#ketinggian maksimum perhitungan numerik
hmax_numerik = np.max(array_y)                                              
print("Ketinggian maksimum perhitungan numerik  : ", hmax_numerik, " m")
```

PLT NUMERIK TANPA HAMBATAN

```python
plt.figure(figsize= (9,5))
plt.plot(array_x, array_y, c='b', label='Numerik Tanpa Hambatan')
plt.axhline(c='black')
plt.axvline(c='black')
plt.title("Gerak Peluru Solusi Numerik Tanpa Hambatan")
plt.ylabel("Ketinggian")
plt.xlabel("Jarak")
plt.legend(loc="upper right")
plt.show()
```

#### Solusi Analitik Tanpa Hambatan

```python
x_an_arr = [0]                                                                         # array untuk menampung perhitungan analitik pada sumbu x (jarak)
y_an_arr = [0]                                                                         # array untuk menampung perhitungan analitik pada sumbu y (ketinggian)

for t in array_t:        
    x_an = v0 * np.cos(sudut_rad) * t                                                  # rumus 9 : jarak analitik : (v(0) cos teta) * t
    y_an = (0.5 * ay * t**2) + (v0 * np.sin(sudut_rad) * t)                            # rumus 10 : ketinggian analitik : (1/2gt^2 + (v(0) sin teta) * t)
    if y_an < 0:
      break
    x_an_arr.append(x_an)                                                              # menampung hasil perhitungan dari rumus jarak analitik ke dalam array
    y_an_arr.append(y_an)                                                              # menampung hasil perhitungan dari rumus ketinggian analitik ke dalam array

print("Hasil Perhitungan Metode Analitik Tanpa Hambatan")
print()

# waktu total perhitungan analitik tanpa hambatan
ttot_analitik = (2 * v0 * np.sin(sudut_rad))/-g * -1                                        # rumus waktu total analitik : 2 * (v(0) sin teta) / -g * -1 (dikali -1 supaya hasilnya tidak negatif)
print("Waktu total perhitungan analitik tanpa hambatan          : ", ttot_analitik, " s")

# jarak maksimum perhitungan analitik tanpa hambatan
rmax_analitik = v0 * np.cos(sudut_rad) * ttot_analitik                                         # rumus jarak maksimum analitik : (v(0) cos teta) / hasil dari waktu total
print("Jarak maksimum perhitungan analitik tanpa hambatan       : ", rmax_analitik, " m")

# ketinggian maksimum perhitungan analitik tanpa hambatan
hmax_analitik = (v0**2 * np.sin(sudut_rad)**2) / (-2 * g) * -1                             # rumus ketinggian maksimum analitik : v(0)^2 sin^2 teta / -2g * -1 (dikali -1 supaya hasilnya tidak negatif)
print("Ketinggian maksimum perhitungan analitik tanpa hambatan  : ", hmax_analitik, " m")
```

PLT ANALITIK TANPA HAMBATAN

```python
plt.figure(figsize= (9,5))
plt.plot(x_an_arr, y_an_arr, c='r', label='Analitik Tanpa Hambatan')
plt.axhline(c='black')
plt.axvline(c='black')
plt.title("Gerak Peluru Solusi Analitik Tanpa Hambatan")
plt.ylabel("Ketinggian")
plt.xlabel("Jarak")
plt.legend(loc="upper right")
plt.show()
```

####Solusi Analitik Dengan Hambatan

```python
# analitik dengan hambatan
vx_air = v0 * np.cos(sudut_rad)                                           # rumus 7 : v(0) cos alfa (menghitung kecepatan awal pada sumbu x)
vy_air = v0 * np.sin(sudut_rad)                                           # rumus 8 : v(0) sin alfa (menghitung kecepatan awal pada sumbu y) 

v_air = math.sqrt((vx_air**2) + (vy_air**2))                              # rumus 13 : akar(vx^2 + vy^2)

ax_air = -(d/m) * (v_air * vx_air)                                        # rumus 11 : -(d/m)vvx 
ay_air = -g - (d/m) * (v_air * vy_air)                                    # rumus 12 : -g-(d/m)vvy

x_air_arr = [0]                                                           # untuk menampung hasil perhitungan pada jarak analitik dengan hambatan udara ke dalam array
y_air_arr = [0]                                                           # untuk menampung hasil perhitungan pada ketinggian analitik dengan hambatan udara ke dalam array

for t in array_t:
   x_air = (0.5 * ax_air * pow(t,2)) + (vx_air * t)                        # rumus 9 : jarak analitik : x(0) + (v(0) cos alfa) * t) + (1/2axt^2)
   y_air = (0.5 * ay_air * pow(t,2)) + (vy_air * t)                        # rumus 10 : ketinggian analitik : y(0) + (v(0) sin alfa) * t) - (1/2ayt^2)
   if y_air < 0 :
     break
   x_air_arr.append(x_air)                                                 # menampung hasil perhitungan dari rumus jarak analitik ke dalam array
   y_air_arr.append(y_air)                                                 # menampung hasil perhitungan dari rumus ketinggian analitik ke dalam array

print("Hasil Perhitungan Metode Analitik Dengan Hambatan")
print()

#waktu total perhitungan analitik tanpa hambatan
ttot_air = array_t[-1]                                                  
print("Waktu total perhitungan numerik          : ", ttot_air, " s")

#jarak maksimum perhitungan analitik tanpa hambatan
rmax_air = x_air_arr[-1]                                                  
print("Jarak maksimum perhitungan numerik       : ", rmax_air, " m")

#ketinggian maksimum perhitungan analitik tanpa hambatan
hmax_air = np.max(y_air_arr)                                              
print("Ketinggian maksimum perhitungan numerik  : ", hmax_air, " m")
```

PLT Analitik Dengan Hambatan


```python
plt.figure(figsize= (9,5))
plt.plot(x_air_arr, y_air_arr, c='g', label='Analitik Dengan Hambatan')
plt.axhline(c='black')
plt.axvline(c='black')
plt.title("Gerak Peluru Solusi Analitik Dengan Hambatan")
plt.ylabel("Ketinggian")
plt.xlabel("Jarak")
plt.legend(loc="upper right")
plt.show()
```
